# Data Visualization

# Elbow Plot

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(data_x)
data_y_pred = kmeans.fit_predict(data_x)
```

```python
kmeans.fit(data_x).score(data_x)
```

```python
nc = range(1,11)

Kmeans = [ KMeans(n_clusters= i) for i in nc]
Kmeans

score = [Kmeans[i].fit(data_x).score(data_x) for i in range(len(Kmeans))]
print(score)
```

```python
import matplotlib.pyplot as plt
```

```python
nc = list(range(1,11))
nc
```

```python
plt.plot(nc , score)
```

```python
import numpy as np
```

```python
score = np.absolute(score)
score
```

```python
plt.figure(figsize=(10 , 10))
plt.plot(nc , score ,  marker='*', color = "r") 
plt.xlabel("Number of Cluster")
plt.ylabel(" Sum Square Distanace")
plt.title("Elbow Plot On data")
plt.grid()
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

# plotting of clusters in visuals

```python
plt.figure(figsize=(10 , 10))
plt.scatter(data_x[data_y_pred == 0, 0], data_x[data_y_pred == 0, 1], c = 'red', label = 'Iris-setosa')
plt.scatter(data_x[data_y_pred == 1, 0], data_x[data_y_pred == 1, 1], c = 'blue', label = 'Iris-versicolour')
plt.scatter(data_x[data_y_pred == 2, 0], data_x[data_y_pred == 2, 1], c = 'green', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
plt.show()
```
